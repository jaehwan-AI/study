```python
import pandas as pd
from sklearn.datasets import load_iris
iris = load_iris()
print(iris.DESCR)
```

```python
data = pd.DataFrame(iris.data, columns=iris.feature_names)
s = pd.Series(iris.target, dtype='category')
s = s.cat.rename_categories(iris.target_names)
data['species'] = s
data
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

scale_cols = ['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)', 'petal width (cm)']
data_scaled = scaler.fit_transform(data[scale_cols])
X = data_scaled
print(X.shape)
```

```python
import tensorflow as tf
from sklearn.preprocessing import LabelEncoder
la = LabelEncoder()

Y = la.fit_transform(data['species'])
Y_encoded = tf.keras.utils.to_categorical(Y)
print(Y_encoded)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_re, y_train, y_re = train_test_split(X, Y_encoded, shuffle=True, test_size=0.2)
x_val, x_test, y_val, y_test = train_test_split(x_re, y_re, shuffle=True, test_size=0.5)
print(x_train.shape)
print(x_val.shape)
print(x_test.shape)
print(y_train.shape)
print(y_val.shape)
print(y_test.shape)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

model = Sequential()
model.add(Dense(20, activation='relu', input_shape=(4,)))
model.add(Dense(10, activation='relu'))
model.add(Dense(3, activation='softmax'))

model.summary()
```

```python
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
history = model.fit(x_train, y_train, validation_data=(x_val, y_val), epochs=100, batch_size=1, verbose=1)
```

```python
evaluate = model.evaluate(x_test, y_test, batch_size=1, verbose=2)
evaluate
```
